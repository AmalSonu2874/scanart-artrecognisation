```python
from google.colab import drive
drive.mount('/content/drive')

import os
from PIL import Image
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, array_to_img, load_img
from tqdm import tqdm

# Input and output paths
input_base = "/content/drive/MyDrive/IKARA/Indian_Paintings_Dataset_OG"
output_base = "/content/drive/MyDrive/IKARA/Indian_Paintings_Dataset"

os.makedirs(output_base, exist_ok=True)

# Target dimensions
target_size = (128, 128)
min_images_per_class = 300

# Augmentation generator
datagen = ImageDataGenerator(
    rotation_range=25,
    width_shift_range=0.15,
    height_shift_range=0.15,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

# Process each class
for class_name in os.listdir(input_base):
    class_input_path = os.path.join(input_base, class_name)
    class_output_path = os.path.join(output_base, class_name)
    os.makedirs(class_output_path, exist_ok=True)

    if not os.path.isdir(class_input_path):
        continue

    images = [f for f in os.listdir(class_input_path) if f.lower().endswith(('jpg', 'jpeg', 'png'))]
    img_count = 0

    print(f"\nðŸŽ¨ Augmenting class: {class_name}")

    for idx, filename in enumerate(tqdm(images, desc=f"{class_name}")):
        img_path = os.path.join(class_input_path, filename)

        try:
            img = load_img(img_path, target_size=target_size)
            x = img_to_array(img)
            x = np.expand_dims(x, axis=0)

            # Save original (first version)
            base_name = f"{class_name[:4]}{str(idx).zfill(2)}_00.png"
            img.save(os.path.join(class_output_path, base_name))
            img_count += 1

            # Augment until we reach at least min_images_per_class
            aug_iter = datagen.flow(x, batch_size=1)
            version = 1
            while img_count < min_images_per_class:
                aug_img = next(aug_iter)[0].astype('uint8')
                aug_img_pil = array_to_img(aug_img)
                new_filename = f"{class_name[:4]}{str(idx).zfill(2)}_{str(version).zfill(2)}.png"
                aug_img_pil.save(os.path.join(class_output_path, new_filename))
                img_count += 1
                version += 1

        except Exception as e:
            print(f"âš ï¸ Error processing {filename}: {e}")

    print(f"âœ… {class_name}: {img_count} images saved in {class_output_path}")

print("\nðŸŽ‰ All classes augmented and saved successfully!")
```
