```python
import os
from PIL import Image
import imagehash
from tqdm import tqdm
import numpy as np

# Path to augmented dataset
base_path = "/content/drive/MyDrive/IKARA/Indian_Paintings_Dataset"
target_size = (128, 128)
hash_size = 16

# Loop through each class folder
for class_name in os.listdir(base_path):
    class_path = os.path.join(base_path, class_name)
    if not os.path.isdir(class_path):
        continue

    print(f"\nğŸ§¹ Cleaning class: {class_name}")
    seen_hashes = set()
    duplicates = 0
    updated = 0

    for filename in tqdm(os.listdir(class_path), desc=f"{class_name}"):
        file_path = os.path.join(class_path, filename)

        if not filename.lower().endswith(('jpg', 'jpeg', 'png')):
            os.remove(file_path)
            continue

        try:
            # Open image safely
            with Image.open(file_path) as img:
                img = img.convert('RGB')
                img = img.resize(target_size)

                # Compute perceptual hash
                img_hash = imagehash.average_hash(img, hash_size=hash_size)

                if img_hash in seen_hashes:
                    os.remove(file_path)
                    duplicates += 1
                    continue

                seen_hashes.add(img_hash)

                # Save as PNG (normalized name)
                new_filename = os.path.splitext(filename)[0] + ".png"
                new_path = os.path.join(class_path, new_filename)
                img.save(new_path, "PNG")
                if new_path != file_path:
                    os.remove(file_path)
                updated += 1

        except Exception as e:
            print(f"âš ï¸ Skipping corrupted file {filename}: {e}")
            os.remove(file_path)

    print(f"âœ… {class_name}: {updated} images kept, {duplicates} duplicates removed.")

print("\nğŸ¨ Dataset fully cleaned and standardized!")
```
